```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
table_2023 = pd.read_csv('2023.csv')
table_2022 = pd.read_csv('2022.csv')
table_2021 = pd.read_csv('2021.csv')
table_2019 = pd.read_csv('2019.csv')
table_2018 = pd.read_csv('2018.csv')
table_2017 = pd.read_csv('2017.csv')
table_2016 = pd.read_csv('2016.csv')
table_2015 = pd.read_csv('2015.csv')
table_2014 = pd.read_csv('2014.csv')
table_2013 = pd.read_csv('2013.csv')
table_2012 = pd.read_csv('2012.csv')
```

```python
def find_name(name, df):
    for i in range(len(df)):
        if df.iloc[i, 0] == name:
            return i
    return False

def find_name_two(name, df):
    for i in range(len(df)):
        if df.iloc[i, 1] == name:
            return i
    return False

def assign_difference_fip(df, prev_df):
    diff_arr = []
    name_arr = []
    for i in range(len(df)):
        if df.iloc[i, 0] in prev_df.values:
            diff_arr.append(df.iloc[i, 6] - prev_df.iloc[find_name(df.iloc[i, 0], prev_df), 6])
            name_arr.append(df.iloc[i, 0])
    data = {"Name": name_arr,
            "FIP Diff": diff_arr}
    return pd.DataFrame(data)

def assign_difference_median_secs(df, prev_df):
    diff_arr = []
    name_arr = []
    for i in range(len(df)):
        if df.iloc[i, 1] in prev_df.values:
            diff_arr.append(df.iloc[i, 7] - prev_df.iloc[find_name_two(df.iloc[i, 1], prev_df), 7])
            name_arr.append(df.iloc[i, 1])
    data = {"Name": name_arr,
            "Secs Diff": diff_arr}
    return pd.DataFrame(data)
```

```python
def plot_arr(col_name):
    avg_arr = [table_2023[col_name].sum() / table_2023["Pitches"].sum(), table_2022[col_name].sum() / table_2022["Pitches"].sum(), table_2021[col_name].sum() / table_2021["Pitches"].sum(), table_2019[col_name].sum() / table_2019["Pitches"].sum(), table_2018[col_name].sum() / table_2018["Pitches"].sum(), table_2017[col_name].sum() / table_2017["Pitches"].sum(), table_2016[col_name].sum() / table_2016["Pitches"].sum(), table_2015[col_name].sum() / table_2015["Pitches"].sum(), table_2014[col_name].sum() / table_2014["Pitches"].sum(), table_2013[col_name].sum() / table_2013["Pitches"].sum(), table_2012[col_name].sum() / table_2012["Pitches"].sum()]
    return avg_arr
```
